```python
#Helps us in finding the null values in the dataset

df.isnull().sum()
```

```python
#Replacing the null values with the median as mean can be affected by the outliers. 
#Median is robust to outliers.

df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x:x.fillna(x.median()))
```

```python
df.isnull().sum() #Here null values in the Review Rating are replaced with the median and it is done groupby
```

```python
#changing the column names into snake case
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ','_')
df = df.rename (columns ={'purchase_amount_(usd)':'purchased_amount'})
```

```python
df.columns #We can see all the column names are changed to lowercase
```

```python
#Creating a new column age_group
labels = ['Young Adult', 'Adult', 'Middle_Age', 'Senior']
df['age_group'] = pd.qcut(df['age'], q=4, labels = labels)
```

```python
df[['age','age_group']].head(10)  #Here we can see a  new column of age_group and they are divided into 4 categories
```

```python
#Create a new column purchase_frequency_days
#We created a dictionary assiging the text to numeric values and later replace the text with the numeric values
frequency_mapping ={
    'Weekly': 7,
    'Bi-Weekly': 14,
    'Fortnightly': 14,
    'Monthly': 30,
    'Quarterly': 90,
    'Every 3 months': 90,
    'Annually': 365
}

df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```

```python
df[['purchase_frequency_days','frequency_of_purchases']].head(10)
```

```python
#As above both columns are same we can remove one so we removed promo_code_applied.
df = df.drop('promo_code_used', axis =1)
```
